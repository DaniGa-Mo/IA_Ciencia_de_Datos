### **2da Pregunta:** ¬øCu√°l fue la facturaci√≥n por categoria em 2023?

**Observaci√≥n:** Para la facturaci√≥n, consideraremos aqu√≠ el producto entre el precio unitario y la cantidad, sumados al costo de env√≠o de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gr√°fico de barras horizontales que muestre la facturaci√≥n por categor√≠a de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categor√≠as de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de env√≠o total de la compra.

Deseamos crear una visualizaci√≥n que muestre la facturaci√≥n de cada categor√≠a, donde la facturaci√≥n se calcula multiplicando el precio unitario por la cantidad y sumando el costo de env√≠o.

Utiliza las bibliotecas de visualizaci√≥n de datos de Python para crear esta visualizaci√≥n.


````python
import pandas as pd
import matplotlib.pyplot as plt

# üìå Calcular facturaci√≥n por compra
df_completo['facturacion'] = (df_completo['precio_unitario'] * df_completo['cantidad']) + df_completo['envio']

# üìå Agrupar facturaci√≥n por categor√≠a
facturacion_categoria = (
    df_completo.groupby('categoria')['facturacion']
    .sum()
    .sort_values()
)

facturacion_categoria
````

````python
# Creaci√≥n del Gr√°fico
plt.figure(figsize=(9, 6))
plt.barh(
    facturacion_categoria.index,
    facturacion_categoria.values
)

plt.title('Facturaci√≥n por categor√≠a de producto')
plt.xlabel('Facturaci√≥n total')
plt.grid(axis='x')
# Visualizaci√≥n del Gr√°fico
plt.show()
````

* metodo_pago: con el m√©todo de pago elegido en la compra

````python
def ventas_mensuales():
  # ===============================
  # Paleta Zoop (hex)
  # ===============================
  ROJO_1, ROJO_2, ROJO_3 = "#e23155", "#cc2c4e", "#b32742"
  AZUL_1, AZUL_2, AZUL_3 = "#203f75", "#1c3867", "#19325b"
  BLANCO, GRIS_1, GRIS_2, GRIS_3, GRIS_4, GRIS_5 = (
      "#ffffff", "#ebebeb", "#d9d9d9", "#cccccc", "#555655", "#231f20"
  )
  AQUA_1, AQUA_2, AQUA_3 = "#addcd4", "#9fccc5", "#96bfb9"

  # ===============================
  # Diccionario meses (abreviado)
  # ===============================
  meses = {
      'January': 'Ene',
      'February': 'Feb',
      'March': 'Mar',
      'April': 'Abr',
      'May': 'May',
      'June': 'Jun',
      'July': 'Jul',
      'August': 'Ago',
      'September': 'Sep',
      'October': 'Oct',
      'November': 'Nov',
      'December': 'Dic'
  }

  # ===============================
  # Datos: ventas mensuales
  # ===============================
  df_completo["mes"] = df_completo["fecha"].dt.strftime("%B")

  ventas_mensuales = (
      df_completo.groupby("mes")["facturacion"]
      .sum()
      .reindex(meses.keys())
  )

  # X num√©rico (0..11) para poder dibujar en posiciones 4,7,10,11
  x_pos = np.arange(len(ventas_mensuales))
  x_labels = [meses[m] for m in ventas_mensuales.index]

  # ===============================
  # Figura y ejes (identidad Zoop)
  # ===============================
  plt.figure(figsize=(12, 6))
  fig = plt.gcf()
  ax = plt.gca()

  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # L√≠nea AZUL_1
  plt.plot(x_pos, ventas_mensuales.values, marker="o", color=AZUL_1, linewidth=2)

  # T√≠tulo
  plt.title("Ventas totales mensuales en 2023", fontsize=18, color=GRIS_5, pad=15)

  # Eje X: etiquetas abreviadas
  plt.xticks(x_pos, x_labels, fontsize=12, color=AZUL_1)

  # Eje Y: formato MXN$ valor M (2 decimales) + color/tama√±o
  def formato_mxn_m(x, pos):
      return f"MXN$ {x/1_000_000:.2f} M"

  ax.yaxis.set_major_formatter(FuncFormatter(formato_mxn_m))
  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)

  # Suprimir labels de ejes
  plt.xlabel("")
  plt.ylabel("")

  # Quitar bordes superior y derecho
  ax.spines["top"].set_visible(False)
  ax.spines["right"].set_visible(False)

  # ===============================
  # A√±adir c√≠rculos numerados (como en tu ejemplo)
  # ===============================
  # Puntos 5¬∞, 8¬∞, 11¬∞, 12¬∞ => √≠ndices 4, 7, 10, 11
  for i, point in enumerate([
      (4,  ventas_mensuales.iloc[4]),
      (7,  ventas_mensuales.iloc[7]),
      (10, ventas_mensuales.iloc[10]),
      (11, ventas_mensuales.iloc[11])
  ], start=1):

      # C√≠rculo (marcador circular sin relleno)
      plt.scatter(point[0], point[1], s=200, color="none", edgecolors=ROJO_1, alpha=0.7)

      # N√∫mero a la izquierda del c√≠rculo
      plt.text(point[0] - 0.3, point[1], str(i), fontsize=12, color=ROJO_1,
              ha="right", va="center")

  # Final
  plt.grid(False)
  plt.tight_layout()
  return plt
````

Utiliza las bibliotecas de visualizaci√≥n de datos de Python para crear esta visualizaci√≥n.


````python
def facturacion_por_categoria():
  # ===============================
  # Paleta Zoop (hex)
  # ===============================
  ROJO_1, ROJO_2, ROJO_3 = "#e23155", "#cc2c4e", "#b32742"
  AZUL_1, AZUL_2, AZUL_3 = "#203f75", "#1c3867", "#19325b"
  BLANCO, GRIS_1, GRIS_2, GRIS_3, GRIS_4, GRIS_5 = (
      "#ffffff", "#ebebeb", "#d9d9d9", "#cccccc", "#555655", "#231f20"
  )
  AQUA_1, AQUA_2, AQUA_3 = "#addcd4", "#9fccc5", "#96bfb9"

  # ===============================
  # 1) Calcular facturaci√≥n por compra
  # ===============================
  df_completo["facturacion"] = (df_completo["precio_unitario"] * df_completo["cantidad"]) + df_completo["envio"]

  # ===============================
  # 2) Agrupar facturaci√≥n por categor√≠a y ordenar (mayor a menor)
  # ===============================
  facturacion_categoria = (
      df_completo.groupby("categoria")["facturacion"]
      .sum()
      .sort_values(ascending=False)
  )

  # ===============================
  # 3) Porcentaje: 1¬™ barra / total
  # ===============================
  porcentaje = (facturacion_categoria.iloc[0] / facturacion_categoria.sum()) * 100

  # ===============================
  # 4) Figura (10, 6) y estilo Zoop
  # ===============================
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Barras horizontales AQUA_1
  barras = ax.barh(
      y=facturacion_categoria.index,
      width=facturacion_categoria.values,
      color=AQUA_1
  )

  # Invertir para que la mayor quede arriba (estilo "ranking")
  ax.invert_yaxis()

  # ===============================
  # 5) T√≠tulo
  # ===============================
  ax.set_title("Facturaci√≥n por categor√≠a de producto en 2023", fontsize=18, color=GRIS_5, pad=15)

  # ===============================
  # 6) Ejes: suprimir xticks, xlabel, ylabel; yticks con estilo
  # ===============================
  ax.set_xlabel("")
  ax.set_ylabel("")
  ax.set_xticks([])  # suprime etiquetas y ticks del eje x

  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)
  ax.tick_params(axis="x", length=0)

  # Suprimir TODOS los bordes
  for spine in ax.spines.values():
      spine.set_visible(False)

  # ===============================
  # 7) Etiquetas de valores al lado derecho de cada barra (AZUL_1, size 12)
  # ===============================
  # Margen extra para que quepan los textos a la derecha
  ax.set_xlim(0, facturacion_categoria.max() * 1.15)

  for barra, valor in zip(barras, facturacion_categoria.values):
      ax.text(
          valor + facturacion_categoria.max() * 0.02,          # un poco a la derecha de la barra
          barra.get_y() + barra.get_height() / 2,               # centrado verticalmente
          f"{valor:,.0f}",                                      # formateo con separador de miles
          va="center",
          ha="left",
          fontsize=12,
          color=AZUL_1
      )

  # ===============================
  # 8) Texto dentro del √°rea del gr√°fico (x=0.4e7, y=5)
  # ===============================
  ax.text(
      0.4e7,
      5,
      f"$\\bf{{{porcentaje:.2f}}}$% de las ventas corresponden √∫nicamente a las ventas de Electr√≥nica.\n"
      "Esto equivale a casi el doble en relaci√≥n con todas las dem√°s categor√≠as\n"
      "presentes en el portafolio de Zoop.",
      fontsize=12,
      color=AZUL_3
  )

  # ===============================
  # Final
  # ===============================
  ax.grid(False)
  plt.tight_layout()
  return plt
````
